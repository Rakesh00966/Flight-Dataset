</div>


```python
color = '#0077b6'

def plot_hist(column,bins,title,xlabel,fontsize = 9,rotation = 0):
    plt.figure(figsize =(15,6))
    counts,bins,patches = plt.hist(column,bins =bins,color = color, edgecolor = 'white')
    plt.xlabel(xlabel,fontsize=12)
    plt.title(title,fontsize=15)
    plt.ylabel("Frequency",fontsize=12)
    
    bin_centers = 0.5*(bins[:-1]+bins[1:])

    for count,x in zip(counts,bin_centers):
        if count>0:
            plt.text(x, count, str(int(count)),fontsize=fontsize,ha='center',  va='bottom',rotation=rotation)
    plt.show()

def plot_bar(column,title,xlabel,colors = sns.color_palette('Set1'),rotation = 0,fontsize= 9):
    plt.figure(figsize=(15,6))
    counts = column.value_counts()
    counts.plot(kind = 'bar',xlabel = xlabel,ylabel = 'Frequency',title = title,color = colors)
    for i,v in enumerate(counts):
        plt.text(i,v,str(int(v)),va = 'bottom',ha = 'center',rotation = rotation)
    plt.show()
    
```

```python
color_scatter = sns.dark_palette("#0077b6", as_cmap =True)
color_box = sns.dark_palette("#0077b6", as_cmap =False)

def plot_scatter(x,y,xlabel,ylabel,title):
    plt.figure(figsize =(15,5))
    plt.scatter(x,y,c= y,cmap = color_scatter,s=2)
    plt.xlabel(xlabel,fontsize = 12)
    plt.ylabel(ylabel,fontsize = 12)
    plt.title(title,fontsize = 15)
    plt.colorbar(label =ylabel)

def plot_box(x,y,xlabel,ylabel,title,rotation = 0,fontsize = 8):
    plt.figure(figsize=(15,5))
    sns.violinplot(x=x,y=y,palette =color_box)
    plt.xlabel(xlabel,fontsize=12)
    plt.xticks(rotation = rotation)
    plt.title(title,fontsize=12)
    plt.ylabel(ylabel,fontsize=15)
    
```
